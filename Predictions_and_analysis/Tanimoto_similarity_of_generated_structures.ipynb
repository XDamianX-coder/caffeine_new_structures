```python
from rdkit import Chem
from rdkit.DataStructs.cDataStructs import TanimotoSimilarity
```

```python
new_smils = []
for smiles in excel['SMILES_From_Tensor_rotation']:
    canon = Chem.MolToSmiles(Chem.MolFromSmiles(smiles), canonical=True) #isomericSmiles=False
    new_smils.append(canon)
```

```python
len(new_smils)
```

```python
drop_duplicates_smi = list(set(new_smils))
```

```python
len(drop_duplicates_smi)
```

```python
new_file = pd.DataFrame(data=drop_duplicates_smi, columns=['new_SMILES'])
```

```python
new_file['CID'] = 0
new_file['PUBCHEM_SMILES'] = 0
```

```python
read_found_structres = excel_1
```

```python
read_found_structres.head()
```

```python
read_found_structres = read_found_structres.loc[read_found_structres['CID'] > 0]
```

```python
read_found_structres.head()
```

```python
read_found_structres = read_found_structres.reset_index()
```

```python
read_found_structres.head()
```

```python
for r, m in enumerate(read_found_structres['PUBCHEM_SMILES']):
    fingerprint_ = Chem.RDKFingerprint(Chem.MolFromSmiles(m))
    for i, element in enumerate(new_file['new_SMILES']):
        fingerprint_gen_str = Chem.RDKFingerprint(Chem.MolFromSmiles(element))
        sim = TanimotoSimilarity(fingerprint_, fingerprint_gen_str)
        if sim == 1:
            new_file['PUBCHEM_SMILES'][i] = m
            new_file['CID'][i] = read_found_structres['CID'][r]
        else:
            pass
```

## Tanomoto_similarity to initial structures

```python
smifile = "../Data/initial_caffeine.xlsx"
data = pd.read_excel(smifile)
data = data['SMILES']

initial_structures_new = list(set(data))
print(len(initial_structures_new))
```

```python
for element in initial_structures_new:
    new_file[element] = 0
```

```python
new_file
```

```python
for element in initial_structures_new:
    finger_1 = Chem.RDKFingerprint(Chem.MolFromSmiles(element))
    for i, smi in enumerate(new_file['new_SMILES']):
        finger_2 = Chem.RDKFingerprint(Chem.MolFromSmiles(smi))
        ta_sim = TanimotoSimilarity(finger_1, finger_2)
        new_file[element][i] = ta_sim
        if ta_sim == 1:
            new_file[element][i] = 100
```
